```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import when

# Initialize Spark session
spark = SparkSession.builder.appName("SCD_Type1").getOrCreate()

# Sample data
data = [
    (1, "GC001", "Active"),
    (2, "GC002", "Active"),
    (3, "GC003", "Expired")
]

columns = ["id", "gift_card_id", "status"]

df = spark.createDataFrame(data, columns)
df.show(5)
```

```python

# Assume the status of GC002 changes to Blocked
updated_data = [(2, "GC002", "Blocked")]

df_update = spark.createDataFrame(updated_data, columns)

# Perform SCD Type 1 (overwrite)
df_scd1 = df.join(df_update, on="gift_card_id", how="left")\
    .select(df["id"], df["gift_card_id"],
            when(df_update["status"].isNotNull(), df_update["status"]).otherwise(df["status"]).alias("status"))

df_scd1.show()
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
from pyspark.sql import SparkSession

# Define the schema explicitly
schema_scd3 = StructType([
    StructField("id", IntegerType(), True),
    StructField("gift_card_id", StringType(), True),
    StructField("status", StringType(), True),
    StructField("previous_status", StringType(), True)  # Explicitly defining the type of previous_status
])

# Sample data with previous status
data_scd3 = [
    (1, "GC001", "Active", None),
    (2, "GC002", "Active", None),
    (3, "GC003", "Expired", None)
]

# Create DataFrame with the explicit schema
df_scd3 = spark.createDataFrame(data_scd3, schema=schema_scd3)

# Assume the status of GC002 changes to Blocked
updated_data_scd3 = [(2, "GC002", "Blocked", "Active")]

# Create DataFrame for updated data with the same schema
df_update_scd3 = spark.createDataFrame(updated_data_scd3, schema=schema_scd3)

# Perform SCD Type 3 (add new column for previous status)
df_scd3_final = df_scd3.join(df_update_scd3, on="gift_card_id", how="left")\
    .select(
        df_scd3["id"],
        df_scd3["gift_card_id"],
        coalesce(df_update_scd3["status"], df_scd3["status"]).alias("status"),
        coalesce(df_update_scd3["previous_status"], df_scd3["previous_status"]).alias("previous_status")
    )

df_scd3_final.show()
```

Question: How do you perform a windowed aggregation to calculate the cumulative sum of transactions for each customer in a retail dataset?

```python
from pyspark.sql import SparkSession
from pyspark.sql.window import Window
from pyspark.sql.functions import sum, col

# Initialize Spark session
spark = SparkSession.builder.appName("WindowAggregation").getOrCreate()

# Sample data representing transactions
data = [
    ("C001", "2024-08-01", 100),
    ("C001", "2024-08-02", 200),
    ("C001", "2024-08-03", 150),
    ("C002", "2024-08-01", 300),
    ("C002", "2024-08-02", 400)
]

columns = ["customer_id", "transaction_date", "transaction_amount"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Define a window specification
window_spec = Window.partitionBy("customer_id").orderBy("transaction_date").rowsBetween(Window.unboundedPreceding, Window.currentRow)

# Calculate cumulative sum using window function
df_with_cumsum = df.withColumn("cumulative_sum", sum(col("transaction_amount")).over(window_spec))

# Display the results
df_with_cumsum.show(truncate=False)

# Explanation:
# - The window specification is defined to partition data by customer_id and order it by transaction_date.
# - rowsBetween(Window.unboundedPreceding, Window.currentRow) defines the window frame to include all previous rows up to the current one.
# - The cumulative sum is then calculated using the sum() function over the specified window.
```

3. Question: How do you use a join operation with complex conditions, including a time-based condition, to find overlapping events for users?

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Initialize Spark session
spark = SparkSession.builder.appName("ComplexJoin").getOrCreate()

# Sample data representing events
events_data = [
    ("U001", "2024-08-01 10:00", "2024-08-01 12:00"),
    ("U001", "2024-08-01 11:00", "2024-08-01 13:00"),
    ("U002", "2024-08-02 09:00", "2024-08-02 11:00"),
    ("U002", "2024-08-02 10:30", "2024-08-02 12:00")
]

columns = ["user_id", "start_time", "end_time"]

# Create DataFrame
df_events = spark.createDataFrame(events_data, columns)

# Self join on complex condition: overlapping time intervals
df_overlapping = df_events.alias("a").join(
    df_events.alias("b"),
    (col("a.user_id") == col("b.user_id")) &
    (col("a.start_time") < col("b.end_time")) &
    (col("a.end_time") > col("b.start_time")) &
    (col("a.start_time") != col("b.start_time")),  # Exclude self-join on the same row
    how="inner"
)

# Display the overlapping events
df_overlapping.show(truncate=False)

# Explanation:
# - A self-join is performed on the 'user_id' to find events belonging to the same user.
# - The conditions (a.start_time < b.end_time) and (a.end_time > b.start_time) ensure that the events overlap in time.
# - The additional condition (a.start_time != b.start_time) prevents a row from joining with itself.
```

4. Question: How do you handle nested data transformations, such as calculating the total price for each order from nested items bought by customers?

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import explode, col, sum

# Initialize Spark session
spark = SparkSession.builder.appName("NestedTransformations").getOrCreate()

# Sample nested JSON data
json_data = '''
[
    {
        "order_id": "ORD001",
        "customer_id": "C001",
        "items_bought": [
            {"item_name": "Laptop", "quantity": 1, "price": 1200.00},
            {"item_name": "Mouse", "quantity": 2, "price": 50.00}
        ]
    },
    {
        "order_id": "ORD002",
        "customer_id": "C002",
        "items_bought": [
            {"item_name": "Tablet", "quantity": 1, "price": 600.00},
            {"item_name": "Headphones", "quantity": 1, "price": 150.00}
        ]
    }
]
'''

# Read JSON data from string
df = spark.read.json(spark.sparkContext.parallelize([json_data]))
df.show()
# Explode the items_bought array
df_exploded = df.withColumn("item", explode("items_bought"))
df_exploded.show()
# Calculate total price for each order
df_total_price = df_exploded.groupBy("order_id", "customer_id").agg(
    sum(col("item.quantity") * col("item.price")).alias("total_price")
)

# Display the DataFrame with total prices
df_total_price.show(truncate=False)

# Explanation:
# - The 'items_bought' field is an array of nested items. explode() is used to flatten this array.
# - After flattening, we calculate the total price for each order by summing the product of quantity and price for each item.
# - groupBy() is used to aggregate the total price at the order level.
```

5. Question: How do you implement a UDF (User-Defined Function) to apply custom transformations on a column, such as masking sensitive data?

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

# Initialize Spark session
spark = SparkSession.builder.appName("UDFTransformation").getOrCreate()

# Sample data representing customers
data = [
    ("C001", "John Doe", "123-456-7890", "john.doe@example.com"),
    ("C002", "Jane Smith", "987-654-3210", "jane.smith@example.com")
]

columns = ["customer_id", "name", "phone", "email"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Define a UDF to mask the email address
def mask_email(email):
    username, domain = email.split("@")
    return f"{username[0]}****@{domain}"

# Register UDF
mask_email_udf = udf(mask_email, StringType())

# Apply the UDF to mask the email column
df_masked = df.withColumn("masked_email", mask_email_udf(col("email")))

# Display the DataFrame with masked emails
df_masked.show(truncate=False)

# Explanation:
# - A User-Defined Function (UDF) is defined to mask part of the email address.
# - The UDF is registered using udf() and applied to the 'email' column to create a new 'masked_email' column.
# - The UDF allows for custom transformations that aren't available through built-in PySpark functions.
```

1. Question: How do you perform a broadcast join to optimize a join operation when one of the DataFrames is small enough to fit in memory?

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import broadcast

# Initialize Spark session
spark = SparkSession.builder.appName("BroadcastJoinExample").getOrCreate()

# Sample data for orders
orders_data = [
    ("ORD001", "C001", 100),
    ("ORD002", "C002", 200),
    ("ORD003", "C003", 150)
]

# Sample data for customers
customers_data = [
    ("C001", "John Doe"),
    ("C002", "Jane Smith"),
    ("C003", "Alice Brown")
]

orders_columns = ["order_id", "customer_id", "amount"]
customers_columns = ["customer_id", "customer_name"]

# Create DataFrames
df_orders = spark.createDataFrame(orders_data, orders_columns)
df_customers = spark.createDataFrame(customers_data, customers_columns)

# Perform a broadcast join
df_broadcast_join = df_orders.join(broadcast(df_customers), on="customer_id", how="inner")

# Display the joined DataFrame
df_broadcast_join.show(truncate=False)

# Explanation:
# - A broadcast join is performed by using the broadcast() function on the smaller DataFrame (customers).
# - This operation optimizes the join by broadcasting the smaller DataFrame to all worker nodes, avoiding a shuffle.
# - Broadcast joins are particularly effective when one DataFrame is small enough to fit in memory.
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Initialize Spark session
spark = SparkSession.builder.appName("BroadcastVariableExample").getOrCreate()

# Create a broadcast variable containing a discount rate
discount_rate = 0.1
broadcast_discount_rate = spark.sparkContext.broadcast(discount_rate)

# Sample data representing product sales
sales_data = [
    ("P001", "Laptop", 1200),
    ("P002", "Smartphone", 800),
    ("P003", "Tablet", 600)
]

columns = ["product_id", "product_name", "price"]

# Create DataFrame
df_sales = spark.createDataFrame(sales_data, columns)

# Apply the broadcast variable in a transformation to calculate the discounted price
df_sales_with_discount = df_sales.withColumn(
    "discounted_price", col("price") * (1 - broadcast_discount_rate.value)
)

# Display the DataFrame with discounted prices
df_sales_with_discount.show(truncate=False)

# Explanation:
# - The broadcast variable is created using sparkContext.broadcast() and holds a read-only value (discount rate).
# - This variable is shared across all tasks without being serialized multiple times, improving performance.
# - The broadcast variable is accessed using .value inside the transformation to calculate discounted prices.
```

3. Question: How do you use an accumulator to count the number of invalid records during a transformation in a Spark job?

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import udf
from pyspark.sql.types import BooleanType

# Initialize Spark session
spark = SparkSession.builder.appName("AccumulatorExample").getOrCreate()

# Create an accumulator to count invalid records
invalid_record_count = spark.sparkContext.accumulator(0)

# Sample data with some invalid records (negative amounts)
data = [
    ("ORD001", 100),
    ("ORD002", -200),  # Invalid record
    ("ORD003", 150),
    ("ORD004", -50)   # Invalid record
]

columns = ["order_id", "amount"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Define a UDF for validation
def validate_record(amount):
    if amount < 0:
        invalid_record_count.add(1)
        return False
    return True

# Register the UDF
validate_record_udf = udf(validate_record, BooleanType())

# Apply the filter using the UDF
df_valid = df.filter(validate_record_udf(col("amount")))

# Trigger an action to execute the transformations
df_valid.count()

# Display the count of invalid records
print(f"Number of invalid records: {invalid_record_count.value}")
```

4. Question: What is the difference between narrow transformations and wide transformations, and how do they impact Spark's performance?

```python
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder.appName("NarrowVsWideTransformations").getOrCreate()

# Sample data representing transactions
data = [
    ("ORD001", "C001", 100),
    ("ORD002", "C002", 200),
    ("ORD003", "C001", 150)
]

columns = ["order_id", "customer_id", "amount"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Narrow Transformation: filter
# - This transformation can be executed in parallel without shuffling data between partitions.
df_filtered = df.filter(df.amount > 100)

# Wide Transformation: groupBy
# - This transformation requires shuffling data across partitions to group records by customer_id.
df_grouped = df_filtered.groupBy("customer_id").sum("amount")

# Trigger actions to execute the transformations
df_filtered.show(truncate=False)
df_grouped.show(truncate=False)

# Explanation:
# - Narrow transformations (like filter) don't require data shuffling across partitions. They are more efficient because they can be processed in parallel within each partition.
# - Wide transformations (like groupBy) require shuffling data between partitions because they involve operations that combine or aggregate data across different partitions.
# - Understanding the difference between narrow and wide transformations is crucial for optimizing Spark jobs, as wide transformations can significantly impact performance due to the cost of shuffling.
```

5. Question: How can you use mapPartitions to apply custom logic to each partition in a DataFrame, such as initializing a database connection for each partition?

```python
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder.appName("MapPartitionsExample").getOrCreate()

# Sample data representing user information
data = [
    ("U001", "John Doe"),
    ("U002", "Jane Smith"),
    ("U003", "Alice Brown")
]

columns = ["user_id", "user_name"]

# Create DataFrame
df = spark.createDataFrame(data, columns)

# Define a function to apply to each partition
def process_partition(partition):
    # Simulate initializing a database connection (e.g., to MySQL)
    # db_connection = init_db_connection()
    results = []
    for row in partition:
        # Simulate processing each row with the database connection
        # processed_row = db_connection.process(row)
        processed_row = (row["user_id"], row["user_name"].upper())  # Example transformation
        results.append(processed_row)
    # Close the database connection (not shown)
    return iter(results)

# Apply mapPartitions to process each partition with custom logic
df_processed = df.rdd.mapPartitions(process_partition).toDF(["user_id", "processed_user_name"])

# Display the processed DataFrame
df_processed.show(truncate=False)

# Explanation:
# - mapPartitions() allows you to apply a function to each partition of an RDD (or DataFrame).
# - This can be useful for operations like initializing a resource (e.g., database connection) per partition rather than per row.
# - In this example, each user's name is converted to uppercase, simulating processing with a database connection.
# - Using mapPartitions is more efficient than using map() when you need to manage resources at the partition level.
```

# Summary
Inner Join: Includes only matching rows from both DataFrames.

Left Join: Includes all rows from the left DataFrame and matching rows from the right DataFrame.

Right Join: Includes all rows from the right DataFrame and matching rows from the left DataFrame.

Full Outer Join: Includes all rows from both DataFrames, with NULLs where there is no match.

Left Semi Join: Includes only rows from the left DataFrame that have matching rows in the right DataFrame.

Left Anti Join: Includes only rows from the left DataFrame that do not have matching rows in the right DataFrame.

Cross Join: Includes every combination of rows from both DataFrames.

1. Scenario: You have two DataFrames, one with transaction data and another with customer details. You need to find all transactions where the customer details are missing.

```python
from pyspark.sql import SparkSession

# Initialize Spark session
spark = SparkSession.builder.appName("MissingCustomerDetails").getOrCreate()

# Sample data for transactions
transactions_data = [
    ("T001", "C001", 150),
    ("T002", "C002", 200),
    ("T003", "C003", 250)
]

# Sample data for customers
customers_data = [
    ("C001", "John Doe"),
    ("C002", "Jane Smith")
]

transactions_columns = ["transaction_id", "customer_id", "amount"]
customers_columns = ["customer_id", "customer_name"]

# Create DataFrames
df_transactions = spark.createDataFrame(transactions_data, transactions_columns)
df_customers = spark.createDataFrame(customers_data, customers_columns)

# Perform a left join to find transactions with missing customer details
df_missing_customers = df_transactions.join(df_customers, on="customer_id", how="left_anti")

# Display the result
df_missing_customers.show(truncate=False)

# Explanation:
# - The left anti join is used to filter out transactions where there is no matching customer in the df_customers DataFrame.
# - This join returns only the transactions with customer_id values that do not have corresponding details in the df_customers DataFrame.
```

2. Scenario: You have a DataFrame of sales data and want to get a list of sales where the amount is higher than the average amount across all sales.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import avg, col

# Initialize Spark session
spark = SparkSession.builder.appName("SalesAboveAverage").getOrCreate()

# Sample data for sales
sales_data = [
    ("S001", 100),
    ("S002", 200),
    ("S003", 300),
    ("S004", 400)
]

sales_columns = ["sale_id", "amount"]

# Create DataFrame
df_sales = spark.createDataFrame(sales_data, sales_columns)

# Calculate the average amount
average_amount = df_sales.agg(avg("amount")).collect()[0][0]

# Filter sales where amount is greater than the average
df_sales_above_avg = df_sales.filter(col("amount") > average_amount)

# Display the result
df_sales_above_avg.show(truncate=False)

# Explanation:
# - First, calculate the average amount across all sales using the avg() function.
# - Then, filter the DataFrame to include only those sales where the amount is greater than the calculated average.
# - This results in a list of sales transactions that exceed the average amount.
```

3. Scenario: You need to find all customers who made purchases in the last month but did not make any purchases in the previous month.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, date_sub, current_date

# Initialize Spark session
spark = SparkSession.builder.appName("CustomersLastMonthOnly").getOrCreate()

# Sample data for purchases with date
purchases_data = [
    ("C001", "2024-07-15"),
    ("C002", "2024-07-25"),
    ("C001", "2024-06-10"),
    ("C003", "2024-07-05"),
    ("C004", "2024-06-30")
]

purchases_columns = ["customer_id", "purchase_date"]

# Create DataFrame
df_purchases = spark.createDataFrame(purchases_data, purchases_columns)

# Convert purchase_date to DateType
df_purchases = df_purchases.withColumn("purchase_date", col("purchase_date").cast("date"))

# Define date range for last month and previous month
last_month_start = date_sub(current_date(), 30)  # Approximate last month start
last_month_end = current_date()
previous_month_end = date_sub(last_month_start, 1)
previous_month_start = date_sub(previous_month_end, 30)  # Approximate previous month start

# Filter purchases from last month
df_last_month_purchases = df_purchases.filter(
    (col("purchase_date") >= last_month_start) & (col("purchase_date") <= last_month_end)
)

# Filter purchases from previous month
df_previous_month_purchases = df_purchases.filter(
    (col("purchase_date") >= previous_month_start) & (col("purchase_date") <= previous_month_end)
)

# Find customers who purchased only in the last month and not in the previous month
df_customers_last_month_only = df_last_month_purchases.join(
    df_previous_month_purchases, on="customer_id", how="left_anti"
)

# Display the result
df_customers_last_month_only.show(truncate=False)

# Explanation:
# - Filter the purchases DataFrame to get transactions from the last month and the previous month.
# - Use a left anti join to find customers who have made purchases in the last month but not in the previous month.
# - This results in customers who are exclusive to the last month's purchases.
```

4. Scenario: You need to calculate the cumulative sum of sales for each month.

```python
from pyspark.sql import SparkSession
from pyspark.sql.window import Window
from pyspark.sql.functions import sum, col, to_date

# Initialize Spark session
spark = SparkSession.builder.appName("CumulativeSumByMonth").getOrCreate()

# Sample data for sales with date
sales_data = [
    ("2024-07-01", 100),
    ("2024-07-15", 200),
    ("2024-08-01", 300),
    ("2024-08-10", 400)
]

sales_columns = ["sale_date", "amount"]

# Create DataFrame
df_sales = spark.createDataFrame(sales_data, sales_columns)

# Convert sale_date to DateType
df_sales = df_sales.withColumn("sale_date", to_date(col("sale_date"), "yyyy-MM-dd"))

# Define a window specification for cumulative sum calculation
window_spec = Window.orderBy("sale_date").rowsBetween(Window.unboundedPreceding, Window.currentRow)

# Calculate cumulative sum of sales
df_sales_cumulative = df_sales.withColumn("cumulative_amount", sum("amount").over(window_spec))

# Display the result
df_sales_cumulative.show(truncate=False)

# Explanation:
# - Convert the sale_date column to DateType to ensure proper ordering.
# - Define a window specification that orders by sale_date and calculates the cumulative sum up to the current row.
# - Apply the sum function over this window specification to calculate the cumulative amount of sales.
# - The resulting DataFrame includes the cumulative sum of sales for each date.
```

5. Scenario: You have a DataFrame with multiple purchase records for each customer. You want to find the total purchase amount for each customer, only including customers who have made more than 3 purchases.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, count, sum

# Initialize Spark session
spark = SparkSession.builder.appName("TotalAmountForFrequentCustomers").getOrCreate()

# Sample data for purchases
purchases_data = [
    ("C001", 100),
    ("C001", 150),
    ("C001", 200),
    ("C001", 50),
    ("C002", 250),
    ("C002", 100),
    ("C003", 300)
]

purchases_columns = ["customer_id", "amount"]

# Create DataFrame
df_purchases = spark.createDataFrame(purchases_data, purchases_columns)

# Calculate the count and total amount of purchases per customer
df_customer_stats = df_purchases.groupBy("customer_id").agg(
    count("amount").alias("purchase_count"),
    sum("amount").alias("total_amount")
)

# Filter customers with more than 3 purchases
df_frequent_customers = df_customer_stats.filter(col("purchase_count") > 3)

# Display the result
df_frequent_customers.show(truncate=False)

# Explanation:
# - Group the purchases DataFrame by customer_id to calculate the total amount and count of purchases per customer.
# - Filter the resulting DataFrame to include only those customers with more than 3 purchases.
# - The result contains the total purchase amount for customers who are considered frequent buyers.
```
